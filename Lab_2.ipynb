```python
import pandas as pd
import numpy as np  
```

```python
def format_lab(data, question, output_mean=True):
    filename = "features/q" + str(question) + ".feature"
    saving = data.reset_index()
    if "index" in saving.keys(): saving = saving.drop('index', 1)
    saving.to_csv(filename, sep=' ', index=False, header=False)
    print "• Data saved on: " + filename
    if(output_mean == True): print "  AVG q%i : %f" % (question, data.mean())
```

```python
yelp_reviews = pd.read_table('yelp_reviews.txt', '|')
```

```python
group_user_id = yelp_reviews.groupby('user_id', sort=True)
```

```python
question_3 = group_user_id.count()['review_id']
format_lab(question_3, 3) 
```

```python
reviews_sum = group_user_id.sum()
format_lab(reviews_sum['votes.cool'], 4) 
format_lab(reviews_sum['votes.funny'], 5) 
format_lab(reviews_sum['votes.useful'], 6) 
```

```python
questions_7 = question_3.apply(np.log).replace(-np.inf, np.nan)
format_lab(questions_7, 7) 
```

```python
log_reviews = reviews_sum.apply(np.log, axis=1).replace([np.inf, -np.inf], np.nan)
log_reviews['votes.cool']
```

```python
percentage_cool = lambda x: x['votes.cool']/(x['votes.cool']+x['votes.funny']+x['votes.useful'])
question_11 = reviews_sum.apply(percentage_cool, axis=1)
format_lab(question_11, 11) 
```

```python
percentage_funny = lambda x: x['votes.funny']/(x['votes.cool']+x['votes.funny']+x['votes.useful'])
question_12 = reviews_sum.apply(percentage_funny, axis=1)
format_lab(question_12, 12) 
```

```python
percentage_useful = lambda x: x['votes.useful']/(x['votes.cool']+x['votes.funny']+x['votes.useful'])
question_13 = reviews_sum.apply(percentage_useful, axis=1)
format_lab(question_13, 13) 
```

```python
yelp_reviews['year'] = yelp_reviews['date'].apply(lambda x: int(x.split('-')[0]))
# We group review by user_id and then the year
# for each user and for each year we can see the reviews

# => the format of years_grouped is: years_grouped[user_id][year] = [review details]
# having user_id and year as the keys of years_grouped

years_grouped = yelp_reviews.groupby(['user_id','year'])

# We reduce the the value of years_grouped[user_id][year] = count_rows to the count of reviews
year_count = years_grouped.count()['review_id']

most_reviewed_year_for_users = {}
# We are going to interate on each row
for key in year_count.keys():
    if key[0] not in most_reviewed_year_for_users or most_reviewed_year_for_users[key[0]]['count'] < year_count[key]:
        most_reviewed_year_for_users[key[0]] = {'year': key[1], 'count': year_count[key]}
        
        
question_15_raw = []
for k in most_reviewed_year_for_users:
    question_15_raw.append({'user_id': k, 'year': int(most_reviewed_year_for_users[k]['year'])})

question_15 = pd.DataFrame(data=question_15_raw)
question_15['year'] = question_15['year'].map(lambda x: x-2005)

format_lab(question_15, 15) 
```

```python
pd_output_year = question_15.copy()
pd_output_year['year'] = pd_output_year['year'].map(lambda x: x+2005)
pd_output_year

question_16 = pd.DataFrame(0, index=question_15['user_id'], columns=range(2005,2014))
for row in pd_output_year.values:
    question_16[row[1]][row[0]] = 1
question_16
    
format_lab(question_16, 16, output_mean=False) 
```
